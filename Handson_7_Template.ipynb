### Specifications:

- The program should accept **integer** entries like 15, 20, 25 for the tip percent.
- The program should validate both user entries. That way, the user can’t crash the program by entering invalid data.
- The program should only accept numbers that are **greater than 0**.
- The program should round results to a maximum of two decimal places.


````python
def main():
    print("Tip Calculator\n")
    x = 0
    y = 0
    while True:
        x = input("Cost of meal:")
        try:
            cost = float(x)
        except ValueError as e:
            print("Must be valid decimal number. Please try again.")
        else:
            if cost <= 0:
                print("Must be greater than 0. Please try again.")
            else:
                break
    while True:
        y = input("Tip Percent:")
        try:
            percent = int(y)
        except ValueError as e:
            print("Must be valid integer. Please try again.")
        else:
            if percent <= 0:
                print("Must be greater than 0. Please try again.")
            else:
                break

    print("\nCost of meal: " + str(round(cost,2)))
    print("Tip percent: " + str(percent)+"%")
    print("Tip amount: " + str(round(cost * (percent / 100), 2)))
    print("Total amount: " + str(round(cost + cost * (percent / 100), 2)))

main()
````

### Specifications:
- This program should read the text file named `wizard_all_items.txt` that contains all the items a wizard can carry.
- When the user selects the walk command, the program should randomly pick one of the items that were read from the text file and give the user the option to grab it.
- The current items that the wizard is carrying should be saved in an inventory file. Make sure to update this file every time the user grabs or drops an item.
- The wizard can only carry four items at a time. For the drop command, display an error message if the user enters an invalid integer or an integer that doesn’t correspond with an item.
- Handle all exceptions that might occur so that the user can’t cause the program to crash. If the all items file is missing, display an appropriate error message and exit the program.
- If the inventory file is missing, display an appropriate error message and continue with an empty inventory for the user. That way, the program will write a new inventory file when the user adds items to the inventory.

````python
import random

def PrintMenu():
    print("\nCOMMAND MENU\nwalk - Walk down the path\nshow - Show all items \ndrop - Drop an item\nexit - Exit program")
    return str(input("Command:"))

def GetAllItems():
    all_items = {}
    i = 1
    with open("wizard_all_items.txt", 'r') as file:
        for line in file:
            all_items[i] = line.rstrip("\n")
            i+=1
    return all_items

def GetInventory():
    inventoy_items = []
    with open("wizard_inventory.txt", 'r') as file:
        for line in file:
            inventoy_items.append(line.rstrip("\n"))
    return inventoy_items

def SaveInventory(inventory_items):
    with open("wizard_inventory.txt", 'w') as file:
        for item in inventory_items:
            file.write(item + "\n")

def InitiateInventory():
    try:
        with open("wizard_inventory.txt", 'r') as file:
            file.close
    except:
        print("\nCould not find inventory file!\nWizard is starting with no inventory.")
        with open("wizard_inventory.txt", 'w') as file:
            file.close
        
def walk(all_items,inventory_items):
    print("\n")
    random_item = all_items[random.randint(1, len(all_items))]
    print(f"While walking down a path, you see {random_item}.")
    pickup = str(input("Do you want to grab it? (y/n):"))
    if len(inventory_items) == 4:
        print("You cant carry anymore items. Drop something first.")
    else:
        inventory_items.append(random_item)
        SaveInventory(inventory_items)
        print(f"You picked up {random_item}.")

def show(inventory_items):
    print("\n")
    i = 1
    for item in inventory_items:
        print(str(i) + ".", item)
        i+=1

def drop(inventory_items):

    if len(inventory_items) <= 0:
        print("You have no items to drop")
    else:
        drop = int(input("\nNumber of item to drop: ")) - 1
        if drop + 1 > len(inventory_items) or drop + 1 <= 0:
            print("Enter a valid item number.")
        else:
            print(f"You dropped {inventory_items[drop]}")
            inventory_items.pop(drop)
            SaveInventory(inventory_items)

def main():
    print("The Wizzard Inventory Program")
    InitiateInventory()
    while True:
        try:
            with open("wizard_all_items.txt", 'r') as file:
                file.close
        except:
            print("\nCould not find items file.\nExiting program. Bye!")
            break
        
        match PrintMenu():
            case "walk":
                walk(GetAllItems(),GetInventory())
            case "show":
                show(GetInventory())
            case "drop":
                drop(GetInventory())
            case "exit":
                break
            case _:
                print("Enter a valid selection")
    print("\nBye!")


main()
````
