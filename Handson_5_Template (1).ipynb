### Specifications:
- Use the attached text file named `wizard_all_items.txt` that contains a list of all the items that a wizard can carry. (in the corresponding directory)
- When the user selects the walk command, the program should read the items from the file, randomly pick one, and give the user the option to grab it.
- Your program should create another file that stores the items that the wizard is carrying (`wizard_inventory.txt`). Make sure to update this file every time the user grabs or drops an item.
- The wizard can only carry **four** items at a time.
- For the drop command, display an error message if the user enters an invalid number for the item.



````python
import random

def PrintMenu():
    print("\nCOMMAND MENU\nwalk - Walk down the path\nshow - Show all items \ndrop - Drop an item\nexit - Exit program")
    return str(input("Command:"))

def GetAllItems():
    all_items = {}
    i = 1
    with open("wizard_all_items.txt", 'r') as file:
        for line in file:
            all_items[i] = line.rstrip("\n")
            i+=1
    return all_items

def GetInventory():
    inventoy_items = []
    with open("wizard_inventory.txt", 'r') as file:
        for line in file:
            inventoy_items.append(line.rstrip("\n"))
    return inventoy_items

def SaveInventory(inventory_items):
    with open("wizard_inventory.txt", 'w') as file:
        for item in inventory_items:
            file.write(item + "\n")

def walk(all_items,inventory_items):
    print("\n")
    random_item = all_items[random.randint(1, len(all_items))]
    print(f"While walking down a path, you see {random_item}.")
    pickup = str(input("Do you want to grab it? (y/n):"))
    if len(inventory_items) == 4:
        print("You cant carry anymore items. Drop something first.")
    else:
        inventory_items.append(random_item)
        SaveInventory(inventory_items)
        print(f"You picked up {random_item}.")

def show(inventory_items):
    print("\n")
    i = 1
    for item in inventory_items:
        print(str(i) + ".", item)
        i+=1

def drop(inventory_items):

    if len(inventory_items) <= 0:
        print("You have no items to drop")
    else:
        drop = int(input("\nNumber of item to drop: ")) - 1
        if drop + 1 > len(inventory_items) or drop + 1 <= 0:
            print("Enter a valid item number.")
        else:
            print(f"You dropped {inventory_items[drop]}")
            inventory_items.pop(drop)
            SaveInventory(inventory_items)

def main():
    print("The Wizzard Inventory Program")
    while True:
        match PrintMenu():
            case "walk":
                walk(GetAllItems(),GetInventory())
            case "show":
                show(GetInventory())
            case "drop":
                drop(GetInventory())
            case "exit":
                break
            case _:
                print("Enter a valid selection")
    print("Bye!")


main()
````

## 7.3 Monthly Sales
Create a program that reads the sales for 12 months from a file and calculates the total yearly sales as well as the average monthly sales. In addition, this program should let the user edit the sales for any month.

### Console:

### Specifications:
- Use the attached text file named CSV file named `monthly_sales.csv` that contains the month and sales data shown above.
- For the edit command, display an error message if the user doesn’t enter a valid three-letter abbreviation for the month.
- When the user edits the sales amount for a month, the data should be saved to the CSV file immediately. That way, no data is lost, even if the program crashes later.
- Round the results of the monthly average to a maximum of 2 decimal digits.

````python
import csv

def CommandMenu():
    print("\nCOMMAND MENU\nmonthly - View monthly sales\nyearly	- View yearly sumary\nedit	- Edit sales for a month\nexit	- Exit program")
    return str(input("Command: "))

def LoadMonthlySales():
    print("")
    monthly_sales_data = {}
    with open("monthly_sales.csv", 'r', newline='') as file:
        for row in csv.reader(file):
            monthly_sales_data[row[0]] = row[1]
    return monthly_sales_data

def monthly(monthly_sales_data):
    for month in monthly_sales_data:
        print(month,'-',monthly_sales_data[month])
    return 0

def yearly(monthly_sales_data):
    total = 0
    for month in monthly_sales_data:
        total += int(monthly_sales_data[month])
    print(f"Yearly Total: {total}.")
    print(f"Monthly Average: {round(total/len(monthly_sales_data), 2)}")
    return 0

def edit(monthly_sales_data):
    month = input("Three Letter Month: ")
    if month not in monthly_sales_data:
        print("Enter a valid three letter month abbreviation")
        return 0
    else:
        monthly_sales_data[month] = int(input("Sales Amount: "))
        with open("monthly_sales.csv", 'w', newline='') as file:
            for month in monthly_sales_data:
                csv.writer(file).writerow([month,(monthly_sales_data[month])])

def main():
    print("Monthly Sales program")
    while True:
        match CommandMenu():
            case "monthly":
                monthly(LoadMonthlySales())
            case "yearly":
                yearly(LoadMonthlySales())
            case "edit":
                edit(LoadMonthlySales())
            case "exit":
                break
            case _:
                print("Enter a valid selection")
    print("Bye")

main()
````
